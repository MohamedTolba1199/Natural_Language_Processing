#Stemming

```python
from nltk.stem import PorterStemmer
```

```python
stemer = PorterStemmer()
words = ['walking','jumps','running','jumping']

for word in words:
  stemmed_word=stemer.stem(word)
  print(f"{word} >>>>>>{stemmed_word}")
```

#Lemitaziation

```python
nltk.download('wordnet')
from nltk.stem import PorterStemmer , WordNetLemmatizer
```

```python
text = "The quick brown foxes jumped over the lazy dogs"

# Stemming
stemer = PorterStemmer()
stemmed_words =[stemer.stem(word) for word in word_tokenize(text)]
print(stemmed_words)

print("-----------------------------------------------------------------------")
# Lemmatization
lentizer = WordNetLemmatizer()
stemmed_words =[lentizer.lemmatize(word) for word in word_tokenize(text)]
print(stemmed_words)

#look for the "lazy"
```

##search pattern

1- search  >>>>>return the first pattern achieve the condition.

2- findall  >>>>>return all the patterns achieve the condition

3- use groups() to print it

```python
# [('2021 Q1', '4.85'), ('2020 Q4', '3')]

text = '''
Tesla's gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
In previous quarter i.e. FY2020 Q4 it was $3 billion.
'''
pattern ='FY(\d{4}.Q[1-4])[^\$]+\$([0-9\.]+)'

result = re.search(pattern, text)
result.groups()
```
